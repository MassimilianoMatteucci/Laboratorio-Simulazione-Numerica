```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os 

os.getcwd()
os.chdir("C:\\Users\\max\\Desktop\\Lab\\Esercitazione1\\Esercizio1.2")
myfile = open("DatiN=1.dat")
x = np.loadtxt("DatiN=1.dat", usecols=(0), delimiter=' ', unpack='true')
y = np.loadtxt("DatiN=1.dat", usecols=(1), delimiter=' ', unpack='true')
z = np.loadtxt("DatiN=1.dat", usecols=(2), delimiter=' ', unpack='true')

n_bins = 100

n, bins, patches = plt.hist(x, n_bins, density=True, range=(-1,1))
plt.xlabel('x')
plt.ylabel('n trhows in bin')
plt.title('Histogram with variables sampled with uniform distribution from -1 to 1 N=1')
plt.grid(True)

plt.figure()
plt.xlabel('x')
plt.ylabel('n trhows in bin')
plt.title('Histogram with variables sampled with exponential distribution N=1')
n, bins, patches = plt.hist(y, n_bins, density=True, facecolor='g', range=(0,5))
plt.grid(True)

plt.figure()
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram with variables sampled with Lorentzian distribution N=1')
n, bins, patches = plt.hist(z, n_bins, density=True, facecolor='r', range=(-5,5))
plt.grid(True)

plt.show()
```
