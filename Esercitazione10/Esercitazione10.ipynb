```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os 
```

```python
Lbest= np.loadtxt("Esercizio10.1//Lengths.dat", usecols=(0), delimiter=' ', unpack='true')
LbestGA= np.loadtxt("Esercizio10.1//LengthsGA.dat", usecols=(0), delimiter=' ', unpack='true')
LmeanGA= np.loadtxt("Esercizio10.1//LengthsGA.dat", usecols=(1), delimiter=' ', unpack='true')
x= np.loadtxt("Esercizio10.1//BestPath.dat", usecols=(1), delimiter=' ', unpack='true')
y= np.loadtxt("Esercizio10.1//BestPath.dat", usecols=(2), delimiter=' ', unpack='true')
xGA= np.loadtxt("Esercizio10.1//BestPathGA.dat", usecols=(1), delimiter=' ', unpack='true')
yGA= np.loadtxt("Esercizio10.1//BestPathGA.dat", usecols=(2), delimiter=' ', unpack='true')
```

```python
fig= plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Citt sparse su circonferenza SA')
plt.xlabel('Generazione')
plt.ylabel('L')
plt.plot(Lbest)
plt.grid(True)

plt.subplot(2,2,2)
plt.title('Citt sparse su circonferenza GA')
plt.xlabel('Generazione')
plt.ylabel('L')
plt.plot(LbestGA,label="Lbest")
plt.plot(LmeanGA,label="Lmean")
plt.legend(loc="upper right")
plt.grid(True)

plt.subplot(2,2,3)
plt.title('Citt sparse su circonferenza Best path SA')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.grid(True)

plt.subplot(2,2,4)
plt.title('Citt sparse su circonferenza Best path GA')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(xGA,yGA)
plt.grid(True)
plt.show()

print('Lunghezza minore SA=',Lbest[-1])
print('Lunghezza minore GA=',LbestGA[-1])
```

Nel SA si ha un solo cromosoma e si propone come nuovo individuo una mutazione del precedente, la mossa viene sempre accettata se il nuovo individuo ha lunghezza minore del precedente invece e ha una lunghezza maggiore si accetta con probabilit $e^{-beta*(Lnew-Lold))}$ con beta l'inverso della temperatura la quale diminuisce col proseguire della simulazione.
