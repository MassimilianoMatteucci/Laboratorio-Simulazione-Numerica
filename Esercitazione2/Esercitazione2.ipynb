Calcoliamo l'integrale $$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$ con il metodo Monte Carlo utilizzando due distribuzioni di probabilit√† differenti: una uniforme ed l'altra con $P(x)=\frac{6}{5}(1-\frac{x^2}{2}) $, la P(x) si ricava dallo sviluppo di Taylor dell'integrando normalizzato sull'intervallo 0-1.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
y1 = np.loadtxt("Esercizio2.1//Dati.dat", usecols=(0), delimiter=' ', unpack='true')
err1 = np.loadtxt("Esercizio2.1//Dati.dat", usecols=(1), delimiter=' ', unpack='true')

x=np.arange(y1.size)
y2 = np.loadtxt("Esercizio2.1//Dati2.dat", usecols=(0), delimiter=' ', unpack='true')
err2 = np.loadtxt("Esercizio2.1//Dati2.dat", usecols=(1), delimiter=' ', unpack='true')
```

```python
fig= plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.errorbar(x,y1,yerr=err1)
plt.title('Distribuzione uniforme')
plt.xlabel('n passi')
plt.ylabel('I')
plt.grid(True)

plt.subplot(1,2,2)
plt.errorbar(x,y2,yerr=err2)
plt.title('Distribuzione P(x)')
plt.xlabel('N Blocchi')
plt.ylabel('I')
plt.grid(True)

plt.show()
```

### <span style="color:green"> Esercizio2 </span>



3D Random Walks partendo dall'origine su griglia cubica e nel continuo

```python
from numba import njit
from scipy.optimize import curve_fit
def f(x,a):
    return a*x**(1./2.)
```

```python
x = np.loadtxt("Esercizio2.2//Dati.dat", usecols=(0), delimiter=' ', unpack='true')
y = np.loadtxt("Esercizio2.2//Dati.dat", usecols=(1), delimiter=' ', unpack='true')
err = np.loadtxt("Esercizio2.2//Dati.dat", usecols=(2), delimiter=' ', unpack='true')
x2 = np.loadtxt("Esercizio2.2//Dati2.dat", usecols=(0), delimiter=' ', unpack='true')
y2 = np.loadtxt("Esercizio2.2//Dati2.dat", usecols=(1), delimiter=' ', unpack='true')
err2 = np.loadtxt("Esercizio2.2//Dati2.dat", usecols=(2), delimiter=' ', unpack='true')
```

```python
fig= plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
p_opt, p_cov = curve_fit(f,x,y)
y_fit = f(x,p_opt[0])
plt.plot(y_fit,color='r') # plotting fitted function
plt.errorbar(x,y,yerr=err)
plt.xlabel('n passi - RW su reticolo cubico')
plt.ylabel('radice del raggio')
plt.grid(True)


plt.subplot(1,2,2)
p_opt, p_cov = curve_fit(f,x2,y2)
y_fit = f(x,p_opt[0])
plt.plot(y_fit,color='r') # plotting fitted function
plt.errorbar(x2,y2,yerr=err2)
plt.xlabel('n passi - RW nel continuo')
plt.ylabel('radice del raggio')
plt.grid(True)

plt.show()
```

Osserviamo che entrambe le curve possono essere ben approssimate ad una funzione del tipo $f(x)=K\sqrt N$, in rosso, andamento di un fenomeno dal carattere diffusivo con anche l'aumentare dell'incertezza sulla media dei raggi al quadrato.
