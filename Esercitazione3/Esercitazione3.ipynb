```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
y1 = np.loadtxt("DatiC.dat", usecols=(0), delimiter=' ', unpack='true')
err1 = np.loadtxt("DatiC.dat", usecols=(1), delimiter=' ', unpack='true')

x=np.arange(y1.size)
y2 = np.loadtxt("DatiC2.dat", usecols=(0), delimiter=' ', unpack='true')
err2 = np.loadtxt("DatiC2.dat", usecols=(1), delimiter=' ', unpack='true')
```

```python
fig= plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.errorbar(x,y1,yerr=err1)
plt.title('European Call-option price calcolando direttamente S(T)')
plt.xlabel('N Blocchi')
plt.ylabel('C')
plt.axhline(14.975790778311286 ,color='r')
plt.grid(True)

plt.subplot(1,2,2)
plt.errorbar(x,y2,yerr=err2)
plt.title('European Call-option price calcolando S(T) in modo discreto')
plt.xlabel('N Blocchi')
plt.ylabel('C')
plt.axhline(14.975790778311286 ,color='r')
plt.grid(True)

plt.show()
```

L'European Call-option price viene compatibile con la soluzione analitica di Blak-Sholes con entrambi gli approcci.
