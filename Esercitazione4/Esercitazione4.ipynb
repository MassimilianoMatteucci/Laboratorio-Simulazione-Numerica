```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os 
from scipy.optimize import curve_fit
```

```python
def f(x,a,b):  # defining the fitting function
    return a*x+b
```

```python
y = np.loadtxt("Esercizio4.1//TempSetting.dat", usecols=(0), delimiter=' ', unpack='true')
y1 = np.loadtxt("Esercizio4.1//TempSetted.dat", usecols=(0), delimiter=' ', unpack='true')
N=y.size
N1=y1.size
x = np.linspace(0,N,N,endpoint=True)
x1 = np.linspace(0,N1,N1,endpoint=True)
```

```python
fig= plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('TempSetting')
plt.xlabel('N passi/10')
plt.ylabel('Temp')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.title('TempSetted')
plt.xlabel('nstep/10')
plt.ylabel('Temp')

p_opt, p_cov = curve_fit(f, x1, y1)
y_fit = f(x1,p_opt[0],p_opt[1])
plt.plot(x1,y_fit) # plotting fitted function

print("optimized parameters [a,b] =")
print(p_opt)

plt.grid(True)
plt.show()
```

Nel grafico a sinistra sono mostrate cinque simulazioni con la funzione per portare il sistema alla temperatura desiderata attiva, la configurazione iniziale è quella con le particelle perfettamente ordinate.
Nel grafico a destra sono mostrate le temperature di due simulazioni con il sistema libero di evolversi quindi con la funzione per la temperatura disattivata, osserviamo che rimangono in media costanti e vicine alla temperatura target di 1.9.

```python
y = np.loadtxt("Esercizio4.1//EpotEq.dat", usecols=(0), delimiter=' ', unpack='true')
y1 = np.loadtxt("Esercizio4.1//EkinEq.dat", usecols=(0), delimiter=' ', unpack='true')
y2 = np.loadtxt("Esercizio4.1//EtotEq.dat", usecols=(0), delimiter=' ', unpack='true')
N=y.size
x = np.linspace(0,N,N,endpoint=True)
```

```python
fig= plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
plt.plot(x,y)
plt.title('EpotEq')
plt.xlabel('N passi/10')
plt.ylabel('Epot')
plt.grid(True)
p_opt, p_cov = curve_fit(f, x, y)
y_fit = f(x,p_opt[0],p_opt[1])
plt.plot(x,y_fit) # plotting fitted function


plt.subplot(1,3,2)
plt.plot(x,y1)
plt.title('EkinEq')
plt.xlabel('N passi/10')
plt.ylabel('Ekin')
p_opt, p_cov = curve_fit(f, x, y1)
y_fit = f(x,p_opt[0],p_opt[1])
plt.plot(x,y_fit) # plotting fitted function

plt.subplot(1,3,3)
plt.plot(x,y2)
plt.title('EtotEq')
plt.xlabel('N passi/10')
plt.ylabel('Etot')
p_opt, p_cov = curve_fit(f, x, y2)
y_fit = f(x,p_opt[0],p_opt[1])
plt.plot(x,y_fit) # plotting fitted function


plt.grid(True)
plt.show()
```

I grafici qui sopra riportano dati da due simulazioni con il sistema libero di evolversi, si può osservare che in media tutte le variabili rimangono costanti nel corso della simulazione ciò unito alle considerazioni sulla temperatura ci permette di affermare che il sistema è in uno stato di equilibrio.

```python
ytemp = 120*np.loadtxt("Esercizio4.3//ArgonSolid//ave_tempSolid.out", usecols=(0), delimiter=' ', unpack='true')
ytemp_err =120* np.loadtxt("Esercizio4.3//ArgonSolid//ave_tempSolid.out", usecols=(1), delimiter=' ', unpack='true')

yp = 120*1.380649*pow(10,-23)*np.loadtxt("Esercizio4.3//ArgonSolid//ave_epotSolid.out", usecols=(0), delimiter=' ', unpack='true')
yp_err =120*1.380649*pow(10,-23)*np.loadtxt("Esercizio4.3//ArgonSolid//ave_epotSolid.out", usecols=(1), delimiter=' ', unpack='true')

yk = 120*1.380649*pow(10,-23)*np.loadtxt("Esercizio4.3//ArgonSolid//ave_ekinSolid.out", usecols=(0), delimiter=' ', unpack='true')
yk_err =120*1.380649*pow(10,-23)*np.loadtxt("Esercizio4.3//ArgonSolid//ave_ekinSolid.out", usecols=(1), delimiter=' ', unpack='true')

yt = 120*1.380649*pow(10,-23)*np.loadtxt("Esercizio4.3//ArgonSolid//ave_etotSolid.out", usecols=(0), delimiter=' ', unpack='true')
yt_err =120*1.380649*pow(10,-23)*np.loadtxt("Esercizio4.3//ArgonSolid//ave_etotSolid.out", usecols=(1), delimiter=' ', unpack='true')
N=yt.size
x = np.linspace(0,N,N,endpoint=True)
```

```python
fig= plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.errorbar(x,ytemp,ytemp_err)
plt.title('Temp ArgonSolid')
plt.xlabel('N simulazione')
plt.ylabel('Temp in K')
plt.grid(True)

plt.subplot(2,2,2)
plt.errorbar(x,yp,yp_err)
plt.title('Epot ArgonSolid')
plt.xlabel('N simulazione')
plt.ylabel('Epot in J')
plt.grid(True)

plt.subplot(2,2,3)
plt.errorbar(x,yk,yk_err)
plt.title('Ekin ArgonSolid')
plt.xlabel('N simulazione')
plt.ylabel('Ekin in J')
plt.grid(True)

plt.subplot(2,2,4)
plt.errorbar(x,yt,yt_err)
plt.title('Etot ArgonSolid')
plt.xlabel('N simulazione')
plt.ylabel('Etot in J')
plt.grid(True)

plt.show()
```
