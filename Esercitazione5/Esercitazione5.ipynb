```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os 
```

Partendo da molto lontano dall'origine

```python
y = np.loadtxt("RtoEq1000.dat", usecols=(0), delimiter=' ', unpack='true')
y1 = np.loadtxt("RtoEq500.dat", usecols=(0), delimiter=' ', unpack='true')
y2 = np.loadtxt("RtoEq100.dat", usecols=(0), delimiter=' ', unpack='true')
y3 = np.loadtxt("RtoEq1.dat", usecols=(0), delimiter=' ', unpack='true')
```

```python
fig= plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.title('R per Psi 1,0,0 partendo da coordinate (1000;1000;1000)')
plt.plot(y)
plt.xlabel('N passo')
plt.ylabel('R')
plt.grid(True)

plt.subplot(2,2,2)
plt.title('Rper Psi 1,0,0 partendo da coordinate (500;500;500)')
plt.plot(y1)
plt.xlabel('N passo')
plt.ylabel('R')
plt.grid(True)

plt.subplot(2,2,3)
plt.title('R per Psi 1,0,0 partendo da coordinate (100;100;100)')
plt.plot(y2)
plt.xlabel('N passo')
plt.ylabel('R')
plt.grid(True)

plt.subplot(2,2,4)
plt.title('R per Psi 1,0,0 partendo da coordinate (1;1;1)')
plt.plot(y3)
plt.xlabel('N passo')
plt.ylabel('R')
plt.grid(True)
plt.show()
```

Dai grafici qui sopra osserviamo che se il punto di partenza  molto distante dal punto di equilibrio la probabilit di accettare una mossa sar molto ridotta, difatti nei due casi pi lontani non abbiamo avuto nessuna accettazione in centomila tentativi.
