```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os 
```

```python
x = np.loadtxt("MetroEneFinal_h=0.0", usecols=(0), delimiter=' ', unpack='true')
y = np.loadtxt("MetroEneFinal_h=0.0", usecols=(1), delimiter=' ', unpack='true')
y_err = np.loadtxt("MetroEneFinal_h=0.0", usecols=(2), delimiter=' ', unpack='true')
x1 = np.loadtxt("GibbsEneFinal_h=0.0", usecols=(0), delimiter=' ', unpack='true')
y1 = np.loadtxt("GibbsEneFinal_h=0.0", usecols=(1), delimiter=' ', unpack='true')
y1_err = np.loadtxt("GibbsEneFinal_h=0.0", usecols=(2), delimiter=' ', unpack='true')
```

```python
#Exact internal energy curve
points=100
T = np.linspace(0.2,2.3,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
```

```python
fig= plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.errorbar(x,y,y_err,linestyle='none',marker='.')
plt.plot(T, e)
plt.title('Ising 1D, internal energy, Metropolis')
plt.xlabel('Temp')
plt.ylabel('U/N')
plt.grid(True)

plt.subplot(1,2,2)
plt.errorbar(x1,y1,y1_err,linestyle='none',marker='.')
plt.plot(T, e)
plt.title('Ising 1D, internal energy, Gibbs')
plt.xlabel('Temp')
plt.ylabel('U/N')
plt.grid(True)
plt.show()
```

```python
x = np.loadtxt("MetroHeatCapacityFinal_h=0.0", usecols=(0), delimiter=' ', unpack='true')
y = np.loadtxt("MetroHeatCapacityFinal_h=0.0", usecols=(1), delimiter=' ', unpack='true')
y_err = np.loadtxt("MetroHeatCapacityFinal_h=0.0", usecols=(2), delimiter=' ', unpack='true')
x1 = np.loadtxt("GibbsHeatCapacityFinal_h=0.0", usecols=(0), delimiter=' ', unpack='true')
y1 = np.loadtxt("GibbsHeatCapacityFinal_h=0.0", usecols=(1), delimiter=' ', unpack='true')
y1_err = np.loadtxt("GibbsHeatCapacityFinal_h=0.0", usecols=(2), delimiter=' ', unpack='true')
```

```python
#Exact heat capacity curve
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
```

```python
fig= plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.errorbar(x,y,y_err,linestyle='none',marker='.')
plt.plot(T, heat)
plt.title('Ising 1D, Heat Capacity, Metropolis')
plt.xlabel('Temp')
plt.ylabel('C')
plt.grid(True)

plt.subplot(1,2,2)
plt.errorbar(x1,y1,y1_err,linestyle='none',marker='.')
plt.plot(T, heat)
plt.title('Ising 1D, Heat Capacity, Gibbs')
plt.xlabel('Temp')
plt.ylabel('C')
plt.grid(True)
plt.show()
```

```python
#Exact Susceptibility curve
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
```

```python
fig= plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.errorbar(x,y,y_err,linestyle='none',marker='.')
plt.plot(T, X)
plt.title('Ising 1D, Susceptibility, Metropolis')
plt.xlabel('Temp')
plt.ylabel('$\chi$')
plt.grid(True)

plt.subplot(1,2,2)
plt.errorbar(x1,y1,y1_err,linestyle='none',marker='.')
plt.plot(T, X)
plt.title('Ising 1D, Susceptibility, Gibbs')
plt.xlabel('Temp')
plt.ylabel('$\chi$')
plt.grid(True)
plt.show()
```

```python
x = np.loadtxt("MetroMagFinal.0", usecols=(0), delimiter=' ', unpack='true')
y = np.loadtxt("MetroMagFinal.0", usecols=(1), delimiter=' ', unpack='true')
y_err = np.loadtxt("MetroMagFinal.0", usecols=(2), delimiter=' ', unpack='true')
x1 = np.loadtxt("GibbsMagFinal.0", usecols=(0), delimiter=' ', unpack='true')
y1 = np.loadtxt("GibbsMagFinal.0", usecols=(1), delimiter=' ', unpack='true')
y1_err = np.loadtxt("GibbsMagFinal.0", usecols=(2), delimiter=' ', unpack='true')
```

```python
#Exact Magnetization with h=0.2 curve
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
```

```python
fig= plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.errorbar(x,y,y_err,linestyle='none',marker='.')
plt.plot(T, M)
plt.title('Ising 1D, Magnetization M with h = 0.02, Metropolis')
plt.xlabel('Temp')
plt.ylabel('$M$')
plt.grid(True)

plt.subplot(1,2,2)
plt.errorbar(x1,y1,y1_err,linestyle='none',marker='.')
plt.plot(T, M)
plt.title('Ising 1D, Magnetization M with h = 0.02, Gibbs')
plt.xlabel('Temp')
plt.ylabel('$M$')
plt.grid(True)
plt.show()
```
