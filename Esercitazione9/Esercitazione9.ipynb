```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os 
```

Nel Genetic Algortim la selezione è stata effettuata attribuendo probabilità al cromosoma di essere selezionato pari a $\frac{1}{L^2}$ normalizzata sul totale con L la lunghezza associata al cromosoma. 

```python
Lbest= np.loadtxt("LengthsNoCross.dat", usecols=(0), delimiter=' ', unpack='true')
Lmean = np.loadtxt("LengthsNoCross.dat", usecols=(1), delimiter=' ', unpack='true')

x= np.loadtxt("BestPathNoCross.dat", usecols=(1), delimiter=' ', unpack='true')
y= np.loadtxt("BestPathNoCross.dat", usecols=(2), delimiter=' ', unpack='true')
```

```python
fig= plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.title('Città sparse su circonferenza Senza Crossover ')
plt.xlabel('Generazione')
plt.ylabel('L')
plt.plot(Lbest,label="Lbest")
plt.plot(Lmean,label="Lmean")
plt.legend(loc="upper right")
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Città sparse su circonferenza Best path Senza Crossover')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
Lbest= np.loadtxt("LengthsSNoCross.dat", usecols=(0), delimiter=' ', unpack='true')
Lmean = np.loadtxt("LengthsSNoCross.dat", usecols=(1), delimiter=' ', unpack='true')

x= np.loadtxt("BestPathSNoCross.dat", usecols=(1), delimiter=' ', unpack='true')
y= np.loadtxt("BestPathSNoCross.dat", usecols=(2), delimiter=' ', unpack='true')
```

```python
fig= plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.title('Città sparse dentro quadrato Senza Crossover ')
plt.xlabel('Generazione')
plt.ylabel('L')
plt.plot(Lbest,label="Lbest")
plt.plot(Lmean,label="Lmean")
plt.legend(loc="upper right")
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Città sparse dentro quadrato Best path Senza Crossover')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.grid(True)
plt.show()

print('Lunghezza minore=',Lbest[-1])
```
